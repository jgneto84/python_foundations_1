```python
# example 1 cattle body weight

def estimate_cattle_weight(heart_girth, body_length):
    """ 
    Estimate cattle live weight (lbs) from body measurements.
    Formula: (Heart Girth**2 x Body length) / 300 
    """

    weight = round((heart_girth ** 2 * body_length) / 300)
    return weight

# example 

#print(estimate_cattle_weight(heart_girth = 70, body_length = 78))

# for loop

heart_girths = [70, 71, 73, 75]
body_lengths = [78, 79, 80, 81]

for i in range(len(heart_girths)):
    hg = heart_girths[i]
    bl = body_lengths[i]
    weight = estimate_cattle_weight(hg, bl)
    print(f"Cow {i + 1}: Heart Girth = {hg} cm, Body Length = {bl} cm -> Estimated Weight = {weight} lbs")

for hg, bl in zip(heart_girths, body_lengths):
    weight = estimate_cattle_weight(hg, bl)
    print(f"Heart Girth={hg} cm, Body Length={bl} cm -> Weight = {weight:.3f} lbs")
```

```python
# example 2 - Calculate Feed Conversion Ratio 

def feed_conversion_ratio(feed_intake, weight_gain):
    """ 
    FCR = Feed Conversion Ratio
    FCR = Feed Intake (kg) / Weight Gain (kg)
    Lower values = better efficiency
    """

    fcr = feed_intake / weight_gain
    return fcr 

#print(feed_conversion_ratio(25, 5))

fi = [35, 36, 37]
wg = [5, 6, 7]

for i, j in zip(fi, wg):
    feed_conversionR = feed_conversion_ratio(i, j)
    print(f"Feed intake = {i}, Weight gain = {j} -> FCR = {feed_conversionR:.2f}")
```

```python
# Example 3 - eggs production rate 

def egg_production_rate(total_eggs, number_of_hens):
    """
    Calculates hen-day egg production (%)
    """
    return (total_eggs / number_of_hens) * 100

# print(egg_production_rate(950, 1000))

te = [990, 998]
nh = [1000, 1000]

for i, j in zip(te, nh):
    EPR = egg_production_rate(i, j)
    print(f"i = {i}, j = {j}, Egg production rate = {EPR}")
```

```python
# Example 4 Daily Milk yield per sheep

def average_milk_yield(total_milk, number_of_days, number_of_sheep):
    """
    Returns average milk yield per sheep per day.
    """
    return total_milk / (number_of_days * number_of_sheep)

tm = [300, 325]
nd = [10, 10]
ns = [15, 15]

for i, j, z in zip(tm, nd, ns):
    AMY = average_milk_yield(i, j, z)
    print(i, j, z, AMY)
```

```python
# fish

def specific_growth_rate(initial_weight, final_weight, days):
    """
    Specific Growth Rate (SGR) = ((ln(final) - ln(initial)) / days) * 100
    """
    import math
    sgr = ((math.log(final_weight) - math.log(initial_weight)) / days) * 100
    return sgr

iw = [50, 51]
fw = [120, 122]
days = [30, 30]

for i, j, l in zip(iw, fw, days):
    SGR = specific_growth_rate(i, j, l)
    print(i, j, l, SGR)
```

```python
# function with conditional logic

def classify_body_condition(score):
    """
    Classify beef cattle body condition score
    """
    if score < 3:
        return "Thin"
    elif 3 <= score < 6:
        return "Moderate"
    else:
        return "Fat"

#print(classify_body_condition(4))

body_condition_scores = [3, 4, 5, 6]

for i, score in enumerate(body_condition_scores):
    category = classify_body_condition(score)
    print(f"Animal {i + 1}: BCS = {score} -> {category}")

categories = []

for score in body_condition_scores:
    categories.append(classify_body_condition(score))
print(categories)

print(categories.count("Thin"))
print(categories.count("Moderate"))
print(categories.count("Fat"))
```

```python
# function with default parameters

def water_requirement(species = "cattle", weight = 500):
    """
    Estimate daily water requirement (liters)
    """
    if species == "cattle":
        return weight * 0.08
    elif species == "sheep":
        return weight * 0.1
    elif species == "chicken":
        return weight * 0.2
    else:
        return "enter a valid species"

print(water_requirement(species = "swine", weight = 100))
```

# Scenario

You are analyzing cows in a dairy farm. For each cow, you want to track:

- ID number
- Breed
- Age
- Daily milk production (liters)

We will create a Cow class to represent each cow. Then we will show how to store multiple cows and calculate some statistics. 

Step 1. Creat a class 

```python
class Cow: 
    # class cow defines the classes as Cow - once a class is created you can create objects associated with it
    def __init__(self, cow_id, breed, age, daily_milk):
        # the line above is a special method called constructor - will take parameters to work with
        
        # the attribute assignments 
        self.cow_id = cow_id
        self.breed = breed
        self.age = age
        self.daily_milk = daily_milk

    def show_info(self):
        print(f"Cow ID: {self.cow_id}")
        print(f"Breed: {self.breed}")
        print(f"Age: {self.age}")
        print(f"Daily milk: {self.daily_milk}")

# __init__ runs we create a cow
# self.attribute stores its data 
# show_info() prints the data 
```

Step 2. Create Cow objects

```python
# Objects that have been created 

cow1 = Cow(cow_id = 1, breed = "Holstein", age = 4, daily_milk = 25.4)
cow2 = Cow(cow_id = 2, breed = "Holstein", age = 3, daily_milk = 34)
cow3 = Cow(cow_id = 3, breed = "Holstein", age = 3, daily_milk = 43)
cow4 = Cow(cow_id = 4, breed = "Holstein", age = 4, daily_milk = 32)
```

```python
print(cow1.breed)
print(cow1)
```

Step 3. Use the objects

Step 4. Store many cows & analyze data

# Inheritance in OOP

Inheritance lets you:

    -- Avoid repeating code (DRY principle)
        - All animals have ID and age -> define once in Animal

    -- Add specialized behaviors 
        - cows have milk 
        - sheep have wool 

    -- Model real-world structures 
        - Farm management systems 
        - Herd/flock databases
        - Unified data pipelines 
        - Predictive modeling by species 

```python
class Animal:
    def __init__(self, animal_id, species, age):
        self.animal_id = animal_id
        self.species = species
        self.age = age
        
    def show_info(self):
        print(f"ID: {self.animal_id}")
        print(f"Species: {self.species}")
        print(f"Age: {self.age} years")


class Cow(Animal): # inherits from Animal
    def __init__(self, animal_id, age, daily_milk, breed):
        super().__init__(animal_id, "Cow", age) # call the parental constructor instatiated in the first class 
        self.daily_milk = daily_milk
        self.breed = breed

    def show_info(self):
        super().show_info() # reuse Animal's show_info()
        print(f"Breed: {self.breed}")
        print(f"Daily Milk: {self.daily_milk} liters")
```

```python
cow1 = Cow(animal_id = 1, age = 3, daily_milk = 34, breed = "Holstein")
cow1.show_info()
```

```python
class Animal:
    def __init__(self, animal_id, species, age):
        self.animal_id = animal_id
        self.species = species
        self.age = age
        
    def show_info(self):
        print(f"ID: {self.animal_id}")
        print(f"Species: {self.species}")
        print(f"Age: {self.age} years")

class Sheep(Animal):
    def __init__(self, animal_id, age, wool_kg):
        super().__init__(animal_id, "Sheep", age)
        self.wool_kg = wool_kg

    def show_info(self):
        super().show_info()
        print(f"Wool production: {self.wool_kg} kg/year") 
```

```python
sheep1 = Sheep(animal_id = 1, age = 2, wool_kg = 3.5)
sheep1.show_info()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1. DAILY MILK PRODUCTION DATA (liters)
# rows = cows, columns = days
# -----------------------------
milk = np.array([
    [18, 20, 19, 21, 22, 20, 19],   # Cow 1
    [22, 23, 22, 24, 25, 22, 23],   # Cow 2
    [15, 16, 15, 17, 18, 16, 15],   # Cow 3
    [20, 19, 20, 21, 20, 19, 20],   # Cow 4
    [17, 18, 17, 18, 19, 18, 17]    # Cow 5
])

milk_2 = milk * 2
# -----------------------------
# 2. CALCULATE TOTAL WEEKLY MILK PER COW
# -----------------------------
total_per_cow = np.sum(milk, axis=1)
mean_per_cow = np.mean(milk, axis =1)
std_per_cow = np.std(milk, axis = 1)
median_per_cow = np.median(milk, axis = 1)
total_per_week = np.sum(milk, axis = 0)

print("Total milk per cow (weekly): ", total_per_cow)
print("Mean milk per cow: ", mean_per_cow)
print("Standard deviation per cow: ", std_per_cow)
print("Median per cow: ", median_per_cow)
print("Total milk per week: ", total_per_week)


# -----------------------------
# 3. SIMPLE GRAPH: WEEKLY MILK PRODUCTION PER COW
# -----------------------------
cows = ["Cow 1", "Cow 2", "Cow 3", "Cow 4", "Cow 5"]

plt.figure(figsize=(8, 5))
plt.bar(cows, mean_per_cow)
plt.title("Weekly Milk Production per Cow")
plt.ylabel("Milk (liters)")
plt.xlabel("Cow")
plt.tight_layout()
plt.show()
```
